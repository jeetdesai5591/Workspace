```python
import numpy
import matplotlib.pyplot as plt
%matplotlib inline
T = 100.0
dt = 0.01
N = int (T/dt)+1
t = numpy.linspace(0.0, T, N)
z0 = 100.
v  = 10
zt = 100.
g  = 9.81
u = numpy.array([z0, v])

z = numpy.zeros(N)
z[0] = z0  
for n in range(1,N):
    u = u + dt*numpy.array([u[1], g*(1-u[0]/zt)])
    z[n] = u[0]
    
plt.figure(figsize=(10,4))  
plt.ylim(40,160)            
plt.tick_params(axis='both', labelsize=14) 
plt.xlabel('t', fontsize=14)
plt.ylabel('z', fontsize=14)
plt.plot(t,z, 'k-');
plt.show()
```
